```
import re
def clean_text(sentence):
  cleaned = sentence
  ## Remove numbers/words contain number
  cleaner = lambda x : re.sub('^\d|\.|-|_', '', x)
  #cleaned = re.sub('[^a-zA-Z ]', '', cleaned)
  words = cleaned.split()
  cleaned = ' '.join([cleaner(word) for word in words if len(word) > 1])
  return cleaned
```

###  Embedding gensim

```
train_data = pd.read_csv("/content/gdrive/My Drive/NDSC2019/data/train.csv")
test_data = pd.read_csv("/content/gdrive/My Drive/NDSC2019/data/test.csv")
```

```
train_data.head()
```

```
corpus = train_data['title'].astype(str)
corpus_test = test_data['title'].astype(str)
```

```
corpus_test.shape
```

```
corpus.shape
```

```
corpus = corpus.apply(clean_text)
corpus_test = corpus_test.apply(clean_text)
```

```
corpus
```

```
from gensim.models import Word2Vec
```

```
from tqdm import tqdm

tqdm.pandas()
```

```
train_sentences = list(corpus.progress_apply(str.split).values)
test_sentences = list(corpus_test.progress_apply(str.split).values)
```

```
train_sentences+ test_sentences
```

```
len(train_sentences + test_sentences)
```

```
model = Word2Vec(train_sentences+ test_sentences, 
                 sg=0, 
                 size=300,
                 window=8,
                 workers=4)
```

```
model.wv.save_word2vec_format('/content/gdrive/My Drive/NDSC2019/data/word2vec_300d_combined1.txt')
```
